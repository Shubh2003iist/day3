```python
 import matplotlib.pyplot as plt
 import numpy as np
```

```python
x=np.array([0.5,0.9,0.11,0.7])
y=np.array([4,10,16,25])
```

```python
plt.plot(x,y,marker='.',ms=45,mfc='k',mec='orange',linestyle='dotted',color='red',linewidth=15)
```

```python
plt.xlabel('time')
plt.ylabel('distance')
plt.title('speed')
plt.plot(x)
plt.plot(y)
plt.show()
```

```python
plt.scatter(x,y)
```

```python
plt.bar(x,y)
```

```python
plt.barh(y,x)
```

```python
plt.hist(x,y)
```

```python
plt.pie(x)
```

```python
l=np.array(['red','green','blue','pink'])
plt.pie(x,labels=l)
```

```python
plt.pie(x)
plt.legend(l)
```

```python
plt.pie(y,explode=x)
```
